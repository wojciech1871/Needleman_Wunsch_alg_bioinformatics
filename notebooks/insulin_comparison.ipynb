## Score functions

In my tests I use two variants of score function:
* first function gives 1 for match, -1 for unmatch and -2 for gap
* second function gives 2 for match, -1 for unmatch and -1 for gap

```python
def score1(x, y):
    if x == "-" or y == "-":
        return -2
    elif x == y:
        return 1
    elif x != y:
        return -1
```

```python
def score2(x, y):
    if x == "-" or y == "-":
        return -1
    elif x == y:
        return 2
    elif x != y:
        return -1
```
